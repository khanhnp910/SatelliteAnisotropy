```python
from astropy.io import fits
from astropy import units as u
from astropy.coordinates import SkyCoord
import astropy.coordinates as coord

mtab = fits.open('../data/NearbyGalaxies_Jan2021_PUBLIC.fits')
```

```python
gname = mtab[1].data['GalaxyName']
pmra = mtab[1].data['pmra']
pmdec = mtab[1].data['pmdec']
vrad = mtab[1].data['vh']
ra, dec = mtab[1].data['RA'], mtab[1].data['dec']
dmod =  mtab[1].data['dmod']

for i, name in enumerate(gname): 
    print(f'{name:s}\t\t {ra[i]:s}\t {dec[i]:s}\t {pmra[i]:.2f}\t {pmdec[i]:.2f}\t {vrad[i]:.2f}')
    d = 10.**(1+0.2*dmod[i])
    c = SkyCoord(ra=ra[i], dec=dec[i], unit=(u.hourangle, u.deg), frame='icrs',
                distance=d*u.pc, pm_ra_cosdec=pmra[i]*u.mas/u.yr, pm_dec=pmdec[i]*u.mas/u.yr,
                radial_velocity=vrad[i]*u.km/u.s)
    #print(pmra[i], pmdec[i])
    gc = c.transform_to(coord.Galactocentric)
    print(gc.x, gc.y, gc.z, gc.v_x, gc.v_y, gc.v_z)
```
